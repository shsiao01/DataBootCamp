```python
import pandas as pd
df = pd.read_excel("titanic3.xls")
df.head()
```

```python
totalSurvived = len(df[df['survived']==1])
```

```python
import numpy as np
df['hasfamily'] = np.where((df['sibsp']==0) & (df['parch']==0), 0,1)
```

```python
hasFamily = len(df[df['hasfamily']==1] )
```

```python
df['accomodation'] = np.where(df['cabin'].notnull(), 1, 0)
totalAccomodation =df['accomodation'].count()
totalFare = df['fare'].count()
avgFare=df['fare'].mean()
print(hasFamily)
print(avgFare)
print(totalAccomodation)
print(totalFare)
```

```python
df.corr()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ hasfamily + accomodation', data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ fare + accomodation', data=df).fit()
result.summary()
```

```python
survivied =  len(df[df['survived']==1])
cabin =len(df[df['cabin'].notnull()])
hasfamily = len(df[df['hasfamily']==1] )
aboveAvgFare=len(df[(df['fare']>avgFare)] )
cabinSurvived = len(df[(df['cabin'].notnull()) & (df['survived']==1)])
hasfamilySurvived = len(df[(df['hasfamily']==1) & (df['survived']==1)])
hasfamilySurvived = len(df[(df['hasfamily']==1) & (df['survived']==1)])

p_cabin_family = (cabinSurvived/survivied) * (hasfamilySurvived/survivied)
p_cabin_avgfare = (cabinSurvived/survivied) * (aboveAvgFare/survivied)
```
